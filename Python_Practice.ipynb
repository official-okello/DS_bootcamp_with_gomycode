```python
# Write a Python program in the form of a function that takes two lists as parameters and returns True if the two lists have at least one common element, and False otherwise.

def common_elements(list1, list2):
    for element in list1:
        if element in list2:
            return True
    return False
list1 = [1, 2, 3, 4, 5]
list2 = [5, 6, 7, 8, 9]

print(common_elements(list1, list2))  # Output: True
```

```python
def common_elements(list1, list2):
    return bool(set(list1) & set(list2))  # Intersection of two sets

# Example Usage
list1 = [1, 2, 3, 4, 5]
list2 = [5, 6, 7, 8, 9]

print(common_elements(list1, list2))  # Output: True
```

```python
class Player:
  def __init__(self, name, hp=100):
    self.name = name
    self.hp = hp

  def take_damage(self,damage):
    self.hp = max(0, self.hp - damage)
    return self.hp

  def __str__(self):
    return f'{self.name} : {self.hp} HP'

player1 = Player('Julius')
player1.take_damage(40)
print(player1)
```

```python
#calculate prime numbers

def is_prime(num):
  if num < 2:
    return False
  if num == 2 or num == 3:
    return True
  else:
    for i in range(2, int(num ** 0.5) + 1):
      if num % i == 0:
        return False
  return True

for i in range(1, 51):
  if is_prime(i):
    print(i)
```

```python
# A particular cell phone plan includes 20 minutes of air time and 20 text messages
# for Ksh. 50.00 a month. Each additional minute of air time costs Ksh. 3.00,
# while additional text messages cost Ksh. 2.00 each.
# The entire bill (including the 999 charge) is subject to 16 percent VAT tax.
# Write a program that reads the number of minutes and text messages used in a month from the user.
# Display the base charge, additional minutes charge (if any),
# additional text message charge (if any), tax and total bill amount.
# Only display the additional minute and text message charges if the user incurred costs in these categories.
# Ensure that all of the charges are displayed using 2 decimal places.
#

def additional_charges(minutes, messages):
  extra_charge = 0
  if minutes > 20:
    extra_charge +=  ((minutes - 20) * 3)
  if messages > 20:
    extra_charge += ((messages - 20) * 2)
  return extra_charge

def calculate_tax(bill):
  return bill * 0.16

def total_bill_amount(bill, tax):
  return bill + tax

def main():
  minutes = int(input('Please enter the number of minutes: '))
  messages = int(input('Please enter the number of messages: '))
  base_charge = 50
  additional_charge = additional_charges(minutes, messages)
  tax = calculate_tax(base_charge + additional_charge)
  total_bill = total_bill_amount(base_charge + additional_charge, tax)

  print(f'Base charge: Ksh {base_charge:.2f}')
  if additional_charge > 0:
    print(f'Additional charge: Ksh {additional_charge:.2f}')
  print(f'Tax: Ksh {tax:.2f}')
  print(f'Total bill: Ksh {total_bill:.2f}')

while True:
  main()
  quit = input('Do you want to quit? (y/n): ').lower()
  if quit == 'y':
    break
```

```python
#downloading the webpage
import requests as re
from bs4 import BeautifulSoup
import pandas as pd

page = re.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168")
soup = BeautifulSoup(page.content, 'html.parser')
seven_day = soup.find(id="seven-day-forecast")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(tonight.prettify())

#extract specifics
period = tonight.find(class_="period-name").get_text()
short_desc = tonight.find(class_="short-desc").get_text()
temp = tonight.find(class_="temp").get_text()
print(period)
print(short_desc)
print(temp)

img = tonight.find("img")
desc = img['title']
print(desc)

#extracting all info
period_tags = seven_day.select(".tombstone-container .period-name")
periods = [pt.get_text() for pt in period_tags]
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
descs = [d["title"] for d in seven_day.select(".tombstone-container img")]
print(periods)
print(short_descs)
print(temps)
print(descs)

#combining data into dataframe
weather = pd.DataFrame({
"period": periods,
"short_desc": short_descs,
"temp": temps,
"desc": descs
})

weather.to_csv('weather.csv', index=False)
weather.head()
```

Detecting Outliers

```python
#Using z-scores
import pandas as pd
import numpy as np
from scipy import stats

# Create a DataFrame with sample data
df = pd.DataFrame({'Height': [170, 160, 130, 190, 180, 150, 140, 200, 175, 165, 300]})

# Calculate the z-score of each data point
z_scores = np.abs(stats.zscore(df['Height']))
threshold = 3
outliers = df[z_scores > threshold]

print("Outliers based on z-score:")
print(outliers)
```

```python
#Using IQR
# Calculate Q1, Q3, and IQR
Q1 = df['Height'].quantile(0.25)
Q3 = df['Height'].quantile(0.75)
IQR = Q3 - Q1

# Define threshold and detect outliers
threshold = 1.5
outliers = df[(df['Height'] < (Q1 - threshold * IQR)) | (df['Height'] > (Q3 + threshold * IQR))]

print("Outliers based on IQR:")
print(outliers)
```

Managing Outliers

```python
#Dropping outlers
# Assuming 'outliers' contains the indices of the outliers
df_clean = df.drop(outliers.index)
print("Cleaned DataFrame:")
print(df_clean)
```

```python
# Replace outliers with the median value of the dataset
median_value = df['Height'].median()
df.loc[df.index.isin(outliers.index), 'Height'] = median_value
print("DataFrame with replaced outliers:")
print(df)
```

```python
#Building a dataset
records = []
for result in results:
  date = result.find('strong').text[:-1] + ', 2017'
  lie = result.contents[1][1:-2]
  explanation = result.find('a').text[1:-1]
  url = result.find('a')['href']

  records.append((date, lie, explanation, url))
```
