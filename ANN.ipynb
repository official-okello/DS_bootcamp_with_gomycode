```python
#@title Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
try:
  import tensorflow
except:
  !pip install tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
```

```python
#@title Load the dataset
data = pd.read_csv('/content/Data (2).csv')
data
```

```python
#@title Select features and labels
features = data[['income', 'age', 'loan']].values
labels = data['default'].values.reshape(-1, 1)  # Reshape for OneHotEncoder
```

```python
#@title Initialize OneHotEncoder and transform labels to dense format
encoder = OneHotEncoder()
encoded_labels = encoder.fit_transform(labels).toarray()
```

```python
#@title Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(features, encoded_labels, test_size=0.2, random_state=42)
```

```python
#@title Building the model
model = Sequential([
  Dense(10, input_shape=(3,), activation='relu'),  # Input layer: specify input_shape instead of input_dim
  Dense(2, activation='softmax')  # Output layer
])
```

```python
#@title Compiling the model
optimizer = Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#@title Training the model
model.fit(X_train, y_train, epochs=1000, batch_size=10)
```

```python
#@title Evaluating the model on the test set
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')
```
