# Introduction

Dataset description: This dataset contains historical records for customer purchases from a supermarket X. The objective is to find the association rules to help supermarket owners find new marketing plans to improve their sales.

# EDA

```python
#@title Importing libraries
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import plotly.express as px
```

```python
#@title Dataset
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
              ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
              ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
              ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
              ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
#@title Run the apriori algorithm on the dataset
te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
#@title Confidence
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
#@title Lift
association_rules(frequent_itemsets, metric="lift", min_threshold=1.25)
```

```python
#@title Load dataset
data = pd.read_csv('/content/Market_Basket_Optimisation.csv')
data.head()
```

```python
#@title Info on dataset
data.info()
```

```python
#@title Describe dataset
data.describe(include='all')
```

```python
#@title Nulls
data.isnull().sum()
```

```python
#@title Explore items
items = data.values.flatten()
items = pd.Series(items).dropna().unique()
items[:15] #display 15 items
```

```python
#@title Plots
item_counts = pd.Series(items).dropna().value_counts().reset_index()
item_counts.columns = ["Item", "Count"]

# Plot top 15 items
fig = px.bar(item_counts.head(15), x="Item", y="Count", title="Top 15 Purchased Items")
fig.show()
```

```python
#@title Preprocessing for Apriori
transactions = []
for index, row in data.iterrows():
    transaction = []
    for item in row:
        if pd.notna(item):
            transaction.append(item)
    transactions.append(transaction)

# Create a dataframe where each item is a column
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
#@title Apply Apriori Algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
frequent_itemsets.head()
```

```python
#@title Derive Association Rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
#@title Visualize association rules with plotly
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))

fig = px.scatter(rules, x="confidence", y="lift", hover_data=["antecedents", "consequents"],
                 title="Confidence vs. Lift for Association Rules")
fig.show()
```
