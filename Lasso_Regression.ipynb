```python
#Least Absolute Shrinkage and Selection Operator
#Introduces the L1 penalty on cost funstion which encourages sparsity in the model
#Performs automatic feature selection
#Might reduce a feature's coef to zero effectively removing it if it is not so impactful
```

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

```python
#creating/loading data
X = np.array([[500, 3], [600, 4], [700, 4], [800, 5], [900, 5], [1000, 6], [1100, 6], [1200, 7]])
Y = np.array([150000, 180000, 210000, 240000, 270000, 300000, 330000, 360000])
```

```python
#splitting data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#train the Lasso Regression Model
lasso_model = Lasso(alpha=1.0)
lasso_model.fit(X_train, Y_train)
```

```python
#making predictions
Y_pred = lasso_model.predict(X_test)
```

```python
#analyzing feature selection effect
print("Lasso Regression Coefficients:", lasso_model.coef_)
print("Lasso Regression Coefficients:", lasso_model.intercept_)
```

```python
#visualizating Lasso Regression
plt.scatter(X[:, 0], Y, color='blue', label='Actual Prices')
plt.plot(X[:, 0], lasso_model.predict(X), color='red', linewidth=2, label='Lasso Regression')
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($)")
plt.title("House Price Prediction using Lasso Regression")
plt.legend()
plt.show()
```
