```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('/content/diamonds.csv')
```

Line Chart

```python
avg_price_by_carat = data.groupby('carat')['price'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x='carat', y='price', data=avg_price_by_carat)
plt.title('Average Price of Diamonds by Carat Size')
plt.xlabel('Carat Size')
plt.ylabel('Average Price ($)')
plt.grid(True)
plt.show()
```

Stacked Area Chart

```python
grouped_data = data.groupby(['cut', 'clarity'])['carat'].sum().unstack()

grouped_data.plot(kind='area', stacked=True)
plt.title('Total Carat Weight for Each Diamond Cut, Stacked by Clarity')
plt.xlabel('Cut Quality')
plt.ylabel('Total Carat Weight')
plt.legend(title='Clarity')
plt.show()
```

Bar Chart

```python
avg_price_by_cut = data.groupby('cut')['price'].mean().reset_index()
sns.barplot(x='cut', y='price', data=avg_price_by_cut, palette='Blues')
plt.title('Average Price of Diamonds by Cut Quality')
plt.xlabel('Cut Quality')
plt.ylabel('Average Price ($)')
plt.show()
```

Scatter Plot

```python
plt.scatter(data['carat'], data['price'])
plt.title('Relationship between Carat and Price of Diamonds by Cut Quality')
plt.xlabel('Carat (weight of the diamond)')
plt.ylabel('Price ($)')
plt.show()
```

Histogram

```python
sns.histplot(data['price'], bins=50)
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

Pie Chart

```python
cut_counts = data['cut'].value_counts()
plt.pie(cut_counts, labels=cut_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Diamond Cuts')
plt.show()
```
