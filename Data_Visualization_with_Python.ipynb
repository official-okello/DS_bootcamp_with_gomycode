```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dataset = pd.read_csv('/content/Africa_climate_change.csv')
```

```python
dataset.info()
```

```python
dataset['DATE'] = dataset['DATE'].str[0:8]
dataset['DATE'] = pd.to_datetime(dataset['DATE'])
dataset['PRCP'].fillna(value=dataset['PRCP'].mean(), inplace=True)
dataset['TAVG'].fillna(value=dataset['TAVG'].mean(), inplace=True)
dataset['TMAX'].fillna(value=dataset['TMAX'].mean(), inplace=True)
dataset['TMIN'].fillna(value=dataset['TMIN'].mean(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
# Plot a line chart to show the average temperature fluctuations in Tunisia and Cameroon. Interpret the results.
filtered_dataset = dataset[
    (dataset['COUNTRY'].isin(['Cameroon', 'Tunisia'])) &
    (dataset['DATE'] >= '1980-01-01') &
    (dataset['DATE'] <= '2005-12-31')
]
fig = px.line(filtered_dataset,
              x='DATE', y='TAVG', color='COUNTRY',
              title='Average Temperature Fluctuations in Tunisia and Cameroon')
fig.show()
```

```python
# Create Histograms to show temperature distribution in Senegal between [1980,2000] and [2000,2023] (in the same figure). Describe the obtained result.
filtered_dataset1 = dataset[
    (dataset['COUNTRY'] == 'Senegal') & (dataset['DATE'] >= '1980-01-01') & (dataset['DATE'] <= '2000-12-31')]
filtered_dataset2 = dataset[
    (dataset['COUNTRY'] == 'Senegal') & (dataset['DATE'] >= '2000-01-01') & (dataset['DATE'] <= '2023-12-31')]

filtered_dataset1['PERIOD'] = '1980-2000'
filtered_dataset2['PERIOD'] = '2000-2023'

combined_senegal = pd.concat([filtered_dataset1, filtered_dataset2])

fig = px.histogram(combined_senegal, x='TAVG', color='PERIOD',
                   nbins=30, barmode='overlay',
                   title='Temperature Distribution in Senegal (1980–2000 vs 2000–2023)',
                   labels={'TAVG': 'Average Temperature (°C)', 'PERIOD': 'Time Period'})
fig.show()
```

```python
# Select the best chart to show the Average temperature per country.
avg_temp_per_country = dataset.groupby('COUNTRY')['TAVG'].mean().reset_index()
fig = px.line(avg_temp_per_country,
              x='COUNTRY',
              y='TAVG',
              title='Average Temperature per Country',
              labels={'COUNTRY': 'Country', 'TAVG': 'Average Temperature (°C)'})
fig.show()
```
